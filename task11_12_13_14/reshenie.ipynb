```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt
```

```python
alpha = 0.05
```

```python
n = 139
m = 1000

S1 = n/(n - 1) * 5.722 ** 2
S2 = m/(m - 1) * 6.161 ** 2

dots = np.linspace(1 / 10, 2, 100) # b / a

left_bound = f.ppf(alpha / 2, n - 1, m - 1)
right_bound = f.ppf(1 - alpha / 2, n - 1, m - 1)

vals = f.cdf(left_bound / dots, n - 1, m - 1) + f.sf(right_bound / dots, n - 1, m - 1)
plt.plot(dots, vals)
plt.grid()
```

```python
from scipy.stats import norm
alpha = 0.05

print(norm.ppf(1 - alpha))
```
