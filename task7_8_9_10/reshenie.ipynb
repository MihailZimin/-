```python
def fact(x):
    answ = 1
    for i in range(2, x+1):
        answ *= i
    return answ
```

```python
def calc_probability(params, counts):
    """
    Функция для подсчёта логарифма функции максимального правдоподобия
    """
    teta = params[0]
    if teta <= 0: return 1e10

    n = len(counts)

    probabilities = [teta ** i * np.exp(-teta) / fact(i) for i in range(n)]
    probabilities = np.clip(probabilities, 1e-10, 1)

    diff = 1 - np.sum(probabilities)
    probabilities[-1] += diff


    return -np.sum(counts * np.log(probabilities))

opt_param = minimize(calc_probability, x0=[1], args=(np.array([109, 65, 26])))
print(opt_param)
```

```python
#здесь ищем параметры для нормального распределения
def calc_probability(params, counts):
    """
    ОМПГ
    """
    teta1, teta2 = params
    if teta1 <= 0: return 1e10
    if teta2 <= 0: return 1e10

    n = 100
    bins = 10
    start = np.array([-np.inf, 0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5])
    end = np.array([0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5, np.inf])


    probabilities = norm.cdf(end, loc=teta1, scale=teta2) - norm.cdf(start, loc=teta1, scale=teta2)
    probabilities = np.clip(probabilities, 1e-10, 1)

    return -np.sum(counts * np.log(probabilities))

m = np.array(
    [5, 8, 6, 12, 14, 18, 11, 6, 13, 7]
)
opt_param = minimize(calc_probability, x0=[1, 1], args=(m))
print(opt_param)
```
